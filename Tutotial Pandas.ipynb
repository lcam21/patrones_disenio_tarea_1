**NumPy**
NumPy es utilizad para el manejo de arrays (matices y vectores). Usa objetos de clase ndarray
Para arrays:
    *np.array()* para covertir una lista de python 
    *np.zeros()* np.ones(), np.arange() para crear arrays
    *np.reshape()* para cambiar de vectro a matriz

El ejemplo abajo muestra como se crea un array y la manuputación de este

```python
import numpy as np

#vector de ceros
a = np.zeros(3)
print("a=\n", a)

#matriz de unos
b = np.ones((2,3))
print("b=\n", b)

#vector arange
c = np.arange(10)
print("c=\n", c)

#rechape vector
d = c.reshape((2,5))
print("e=\n",d)

#creación de array
x = np.array ([[1,2,3],[3,1,2],[2,3,1]])
print ("x=\n", x)
```

Existen otras funciones como:
    Multipicación de mmatrices: *x @ y*
    Transpuesta de una matrix: *np.transpose(x)*
    Inversa de una matriz: *nplinalg.inv(x)*

Analizando nuestro dataset:
    Uno de los analisis mas comunes es la distribución de nuestras columnas numericas, como por ejemplo el precio, el año y el kilometraje de los vehiculos

```python
import matplotlib.pyplot as plt

#---- Price ----
plt.suptitle('Histogram')
plt.xlabel('Price [$]')
plt.ylabel('Number of Cars')
df["price"].hist(bins=50)
plt.show()

#---- Year ----
plt.suptitle('Histogram')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
df["year"].hist(bins=50)
plt.show()

#---- Year ----
plt.suptitle('Histogram')
plt.xlabel('Km')
plt.ylabel('Number of Cars')
df["odometer"].hist(bins=50)
plt.show()
```

Otra namera de anlizar los datos es ecplorando la relación entre estos. Para nuestro dataset un buen ejemplo es la relación entre precio vs añoy precio vs kilometraje.

```python
#---- Price vs Year ----
df.plot.scatter(y = ["price"], x = ["year"])
plt.show()


#---- Price vs Odometer ----
df.plot.scatter(y = ["price"], x = ["odometer"])
plt.show()
```
